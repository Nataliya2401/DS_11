```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
fraction = (pd.isnull(fd).sum()/len(fd))
print(fraction)
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
columns_with_empty = fd.columns[fd.isna().any()].tolist()
columns_with_empty.remove('Язык.программирования')
print(columns_with_empty)
# fd1 = fd.dropna
fd1 = fd.drop(columns_with_empty, axis=1)
fd1.head()
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці
fraction1 = (pd.isnull(fd1).sum()/len(fd1))
print(fraction1)
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
fd1 = fd1.dropna()
fd1.head(10)
```

```python
# Визначте новий розмір таблиці за допомогою методу shape
fd1.shape
```

```python
# Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
# python_data = fd1.loc[fd1['Язык.программирования'] == "Python"]  - var1
# fd1 = fd1.rename(columns={"Язык.программирования": "language"}) - var 2
# python_data = fd1.query('language == "Python"')

# query для недопустимого імені викоритовуємо звортні кавички
python_data = fd1.query('`Язык.программирования` == "Python"')
fd1
python_data
```

```python
# Визначте розмір таблиці python_data за допомогою методу shape
python_data.shape
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
group_position = fd1.groupby('Должность')
for Должность, group in group_position:
  print("Посада:   ", Должность)
  print(group.head(3)[['Язык.программирования','Зарплата.в.месяц']], '\n')

# group_position.count()
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
```

```python
group_position_python = python_data.groupby('Должность')["Зарплата.в.месяц"]
group_position_python.agg(['min', 'max'])
```

```python
# Створіть функцію fill_avg_salary яка повертатиме середнє значення заробітної плати на місяць.
# Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(s: pd.Series):
  mean_salary = s['Зарплата.в.месяц'].mean()
  s['avg'] = mean_salary
  return s
```

```python
group_position_all =  fd1.groupby('Должность', as_index=False,  group_keys=True)
fd1 = group_position_all.apply(fill_avg_salary)
fd1
```

```python
#  описова статистика для avg
stat = fd1['avg'].describe()
stat
```

```python
# запис у файл
fd1.to_csv('about_it.csv', index=False)
```
