```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tab_1 = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match="Коефіцієнт народжуваності в регіонах України")

df = tab_1[0]
print(df, '\n')
```

```python
# Вивести перші рядки таблиці за допомогою методу head
print(df.head())
```

```python
# Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
df.shape
```

```python
# Замініть у таблиці значення "—" на значення NaN
df1 = df.replace( "—", np.nan)
print(df1)
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes
df1.dtypes
```

```python
# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
df1.iloc[:, 1:] = df1.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
df1.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
fraction = (pd.isnull(df1)).sum()/len(df1)
print(fraction)
```

```python
# Видаліть з таблиці дані по всій країні, останній рядок таблиці
df1 = df1.drop([27])
# df1.drop(df1.tail(1).index, inplace=True)
print(df1)
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
df1 = df1.fillna(df1.mean().round(2), axis=0)
print(df1)
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
mn  = df1['2019'].mean()
print("Average value in 2019 : ", mn, "\n")
print(df1.loc[df1['2019'] > mn][['Регіон', '2019']])
```

```python
# У якому регіоні була найвища народжуваність у 2014 році?
print(df1.loc[df1['2014'] == df1['2014'].max()][['Регіон', '2014']])
```

```python
#  Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
sns.set_style('darkgrid')
plt.bar(df1['Регіон'], df1['2019'])
plt.xticks(rotation=90)
plt.title ('Народжуваність по регіонах у 2019 році')
plt.xlabel('Regions')
plt.ylabel('Birth Rate')
plt.show()
```

```python
plt.style.use('default')
plt.plot(df1['Регіон'], df1['2000'], marker='.', label='2000', linestyle='-.')
plt.plot(df1['Регіон'], df1['2012'], marker='.', label='2012')
plt.plot(df1['Регіон'], df1['2019'], marker='o', label='2019')
plt.xlabel('Регіон', fontsize='small')
plt.ylabel('Народжуваність')
plt.title('Народжуваність по регіонах України', color="blue")
plt.xticks(rotation=90)
plt.grid()
plt.legend()
plt.show()
```
